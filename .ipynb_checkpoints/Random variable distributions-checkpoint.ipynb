## monthly_premium_auto

```python
sns.histplot(numerical.monthly_premium_auto, bins=15)
```

```python
# outliers : just a few value > 150
```

```python
import numpy as np
```

```python
transformed_MPA = np.log(numerical.monthly_premium_auto)
sns.histplot(transformed_MPA, bins=10)
```

```python
# Still outliers, but we can better see them
```

```python
sns.histplot(numerical.customer_lifetime_value, bins=15)
```

```python
# outliers : just a few value > 40000
# the graph is exponential distribution
```

```python
transformed_clv = np.log(numerical.customer_lifetime_value)
sns.histplot(transformed_clv, bins=10)
```

```python
# this gave us a normal distribution => OK
```
